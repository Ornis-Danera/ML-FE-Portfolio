```python
# --- 1️⃣ Imports and setup ---
import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt

# Warnings and display settings
import warnings
warnings.filterwarnings("ignore")
np.set_printoptions(precision=4, suppress=True)



# --- 3️⃣ Download historical price data ---
start_date = "2005-01-01"  # 20+ years history
end_date = "2025-01-01"

data_full = yf.download(tickers, start=start_date, end=end_date)
data = data_full.xs('Close', axis=1, level=0)

# Suppose your full MultiIndex DataFrame is called data_full
# Step 1: Select only 'Close' prices
close_prices = data_full.xs('Close', axis=1, level=0)  # xs selects a cross-section along a level

# Step 2: Keep only tickers that existed in 2005
tickers = ['XLB','XLE','XLF','XLI','XLK','XLP','XLU','XLV','XLY']
close_prices = close_prices[tickers]

# Step 3: Reset index so 'Date' becomes a column
close_prices = close_prices.reset_index()

# Step 4: (Optional) calculate daily returns
returns = close_prices.set_index('Date').pct_change().dropna()

```

```python
# Simple percentage returns
returns = close_prices.pct_change().dropna()

# Log returns
log_returns = np.log(1 + returns)

# Inspect
print("Log returns head:\n", log_returns.head())
print("\nLog returns summary:\n", log_returns.describe())



# Optional: save
close_prices.to_csv("sector_etf_close_prices.csv")
```
